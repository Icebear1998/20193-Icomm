```
with open('/content/drive/My Drive/Colab Notebooks/nlp-20193/X_mostcommon_processed.txt') as f:
  X_processed = f.read().split('\n')[:-1]
```

```
len(X_processed)
```

```
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(analyzer='word')
vect_representation = vect.fit_transform(X_processed)
vect_representation.shape
```

```
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y_data = le.fit_transform(y_processed)
y_data.shape
```

```
from sklearn.feature_selection import SelectPercentile

# 10 means only include the 10% best features
selector = SelectPercentile(percentile=22.776)

# A feature space with only 10% of the features
X_data = selector.fit_transform(vect_representation, y_data)
X_data.shape
```

```
from imblearn.over_sampling import SMOTE, BorderlineSMOTE

smote = BorderlineSMOTE('all')

X, y= smote.fit_sample(X_data, y_data)
print(X.shape, y.shape)
```

```
Counter(le.inverse_transform(y)).most_common()
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, shuffle=True)

from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score

clf = MultinomialNB()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print('Training size = %d, accuracy = %.2f%%' % \
(X_train.shape[0],accuracy_score(y_test, y_pred)*100))
```

```
from sklearn.linear_model import LogisticRegression   
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, shuffle=True)
       
cl=LogisticRegression().fit(X_train, y_train)

train_predictions = cl.predict(X_train)
val_predictions = cl.predict(X_val)

print("Train accuracy: ", accuracy_score(train_predictions, y_train))
print("Validation accuracy: ", accuracy_score(val_predictions, y_val))
```

```
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)

SVM = svm.SVC(C=1.0, kernel='linear', gamma='auto')
SVM.fit(X_train,y_train)
# predict the labels on validation dataset
predictions_SVM = SVM.predict(X_test)
# Use accuracy_score function to get the accuracy
print("SVM Accuracy Score -> ",accuracy_score(predictions_SVM, y_test)*100)
```
