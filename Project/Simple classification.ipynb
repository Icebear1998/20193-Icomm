```python
## a class for simple process

class NLP(object):
    def __init__(self, text = None, SPECIAL_CHARACTER=None, STOP_WORDS_DIR=None):
        self.text = text
        self.SPECIAL_CHARACTER = SPECIAL_CHARACTER
        self.STOP_WORDS_DIR = STOP_WORDS_DIR
        self._set_stopwords()       
        
    def segmentation(self):
        return ViTokenizer.tokenize(self.text)
    
    def _set_stopwords(self):
        if self.STOP_WORDS_DIR != None:
            text_file = open(self.STOP_WORDS_DIR, "r",encoding="utf8")
            lines = text_file.read().split('\n')
            self.stopwords = [stopword.replace(' ','_') for stopword in lines]
        else:
            self.stopwords = []
        
    def split_words(self):
        text = self.segmentation()
        if self.SPECIAL_CHARACTER != None:
            return [x.strip(self.SPECIAL_CHARACTER).lower() for x in text.split() 
            if x.strip(self.SPECIAL_CHARACTER)!="" and all((c not in self.SPECIAL_CHARACTER) for c in x)]
        else:
            return text

    def get_words_feature(self):
        split_words = self.split_words()
        return [word for word in split_words if word not in self.stopwords]
```

```python
# preprocess text and save in file

SPECIAL_CHARACTER = '0123456789%@$.,=+-!;/()*"&^:#|\n\t\'"“”…>>?<'
STOP_WORDS_DIR = '/content/drive/My Drive/Colab Notebooks/nlp-20193/vietnamese-stopwords.txt'
for i in range(len(X)):
  print("Step {} / {}".format(i, len(X)))
  nlp = NLP(text=X[i],SPECIAL_CHARACTER=SPECIAL_CHARACTER,STOP_WORDS_DIR=STOP_WORDS_DIR)
  X[i] = nlp.get_words_feature()
print(nlp.get_words_feature())

f=open('/content/drive/My Drive/Colab Notebooks/nlp-20193/X_mostcommon_processed.txt','w')

for ele in X:
  for word in ele:
    f.write(word+' ')
  f.write('\n')

f.close()
```
